```python
# Open the PDF file in read-binary mode
with open("honeyproduction 1998-2021.pdf", 'rb') as file:

    # Create a PDF object
    pdf = PyPDF2.PdfFileReader(file)

    for page in range(pdf.getNumPages()):
        # Extract the text from the page
        text = pdf.getPage(4).extractText()

        print(text)
```

```python


# Use a regular expression to extract the phone number from the text
numb_pattern = r'\d{6}' # Here this will extract the number using a regular expression that matches thus pattern of 6 digits here.
numb_number = re.search(numb_pattern, text).group()

print(f'honey production: {numb_number}')
#We got the otput as 434000 here as it was the first number with 6 digits remaining numbers were discarded.
```
